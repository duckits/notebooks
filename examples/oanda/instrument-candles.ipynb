```python
import os
import sys
sys.path.append('../../')
from fxtrade.broker.oanda import Oanda
from datetime import date, timedelta
import json
```

```python
data_dir="../../data"
```

```python
client = Oanda()
```

```python
def get_datelist(year):
    dates = []
    start_date = date(int(year), 1, 1) 
    end_date = date(int(year), 12, 31)
    delta = end_date - start_date

    for i in range(delta.days + 1):
        day = start_date + timedelta(days=i)
        dates.append(day)
    return dates
```

```python
def get_candles_for(instrument, startDate, endDate, granularity='M1', price='MBA'):
    params = {
        'granularity': granularity, 
        'price': price,
        'from': f"{startDate}T00:00:00.00Z",
        'to': f"{endDate}T00:00:00.00Z"
    }
    return client.get_instrument_candles(instrument, params)
```

```python
from pathlib import Path

def write_file(file, data):
    path = Path(file)
    os.makedirs(path.parent.absolute(), exist_ok=True)
    with open(path, "w") as outfile:
        json.dump(data, outfile)
```

```python
years = range(2000,date.today().year+1)
years = [2022]
```

```python
with open(f"{data_dir}/instruments.json", "r") as f:
    data = json.loads(f.read())
instruments = list(data.keys())
```

```python
for instrument in instruments:
    for year in years:
        dates = get_datelist(year)
        for i in range(len(dates)):
            try:  
                data = get_candles_for(instrument, dates[i], dates[i+1])
                count = len(data['candles'])
                print(f"{instrument} {dates[i]} {count}")
                if count > 0:
                    write_file(f"{data_dir}/{instrument}/{year}/{dates[i]}-M1.json", data)
            except IndexError:
                pass
```
